# Punto 1:

- `[4pts]` Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con `with open`
- `[4pts]` Convertir a minúsculas
- `[4pts]` Conservar sólo palabras
- `[4pts]` Tokenizar

```python
# Leyendo el archivo
with open ("D:/03. ESPECIALIZACION/Semestre 2/2. NLP/03 Clase/UNA_SOLITARIA_VOZ_HUMANA.txt",encoding="utf8") as archivo:
    texto = archivo.read()
    print(texto)
```

```python
# Convetir Minusculas
texto = texto.lower()
```

```python
# Import Libreria re
import re
# Conservar sólo palabras
texto = re.sub(r"[\W\d_]+", " ", texto)
# Quitar espacios del principio
texto = texto.strip() 
```

```python
#  Tokenizar
texto_list = texto.split()
texto_list
```

# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
# Importar nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
# Sin palabras vacias
texto_list = [palabra for palabra in texto_list if palabra not in stopwords_sp]
texto_list
```

```python
# Importar collections
import collections
```

```python
# Resultado ¿Cuáles son las 10 palabras no vacias más usadas?
counter=collections.Counter(texto_list)
print(counter.most_common(10))
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

```python
# Importar libreria 
from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")
```

```python
# Stemming del documento
texto_raiz = []

for palabra in texto_list:
    raiz = spanishStemmer.stem(palabra)
    texto_raiz.append(raiz) 

texto_raiz
```

```python
# Resultado ¿Cuáles son las 10 raíces más usadas?
counter_raiz=collections.Counter(texto_raiz)
print(counter_raiz.most_common(10))
```

# BONUS Punto 4:

- `[5pts]` Contar cuántas ocurrencias hay por cada parte de la oración

```python
import es_core_news_sm

doc = []

nlp = es_core_news_sm.load()
for palabra in texto_list:
    x = nlp(palabra)
    doc.append(x)
    oracion = []
for y in doc:
    for w in y:
        x = w.pos_
    oracion.append(x)

counter_oracion=collections.Counter(oracion)
print(counter_oracion.most_common())
```
