```python
import re
import pandas as pd 
from pprint import pprint

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import pyLDAvis.gensim
from gensim.models import LdaModel
from gensim.corpora import Dictionary

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Cargar datos
path = 'D:/01. PC LENOVO/02. ESPECIALIZACION/Semestre 2/2. NLP/09 Clase/reviews_vidjew_es.csv'
data = pd.read_csv(path)
data.head()
```

###  `[12 pts]` Punto 1: Hacer pre-procesamiento del texto

```python

def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = " ".join([palabra for palabra in texto.split() if palabra not in stopwords])
    return texto.split()

data['pp'] = data['review_body'].apply(lambda texto: pre_procesado(texto))

data.head()
```

###  `[13 pts]` Punto 2: Modelo de LDA

```python
# Crear una representaci贸n de los documentos en forma de diccionario
dictionary = Dictionary(data['pp'].values)

# Filtrar palabras muy frecuentes o infrecuentes
dictionary.filter_extremes(no_below=5, no_above=0.5)

corpus = [dictionary.doc2bow(text) for text in data['pp'].values]


# Train the topic model
model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=50)
```

###  `[25 pts]` Punto 3: Visualizaci贸n de LDA

```python
lda_display = pyLDAvis.gensim.prepare(model, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
# pyLDAvis.save_html(lda_display, 'lda.html')
```
